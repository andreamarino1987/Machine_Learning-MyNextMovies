```python
# clustering

from sklearn import cluster, datasets


kmeans = cluster.KMeans(n_clusters = 6)

kmeans.fit(x_num)

pred = kmeans.predict(x_num)

display(pred)
```

```python
# PCA

from sklearn.decomposition import PCA

pca = PCA(n_components = 3)

pca.fit(x_num)

x_pca = pca.transform(x_num)
```

```python
x_pca
```

```python
from sklearn import cluster, datasets


kmeans = cluster.KMeans(n_clusters = 6)

kmeans.fit(x_pca)

pred = kmeans.predict(x_pca)

display(pred)
```

```python
plt.figure(figsize = (10, 10))
plt.scatter(x_pca[:, 1], x_pca[:, 2], c= pred, alpha = 0.5)

plt.show()
```

```python
title = pd.DataFrame(datamovie['title'])
features = pd.DataFrame(x_pca)
prediction = pd.DataFrame(pred)
```

```python
a = title.join(features)
```

```python
b = pd.merge(a, prediction, left_index=True, right_index=True)
```

```python
b.rename(columns = {'0_y' : 'clusters'}, inplace = True)
```

```python
rating = datamovie['rating']
```

```python
b.head()
```

```python
b.drop(columns = ['0_x', 1, 2], axis = 1, inplace= True)
```

```python
cluster = pd.concat([b, rating], axis = 1 )
```

```python
cluster = cluster[['title', 'rating', 'clusters']]
```
