```python
import pandas as pd
```

```python
movies = pd.read_csv('movies.csv')

movies
```

```python
ratings = pd.read_csv('ratings.csv')
ratings = ratings.drop('timestamp', axis=1)
ratings
```

```python
links = pd.read_csv('links.csv')

#links
```

```python
tags = pd.read_csv('tags.csv')
tags = tags.drop('timestamp', axis=1)

tags.loc[tags['movieId'] == 1]
```

```python
genome_tags =pd.read_csv('genome-tags.csv')

genome_tags.rename(columns = {'tag' : 'genome'}, inplace = True)

genome_tags
```

```python
genome_scores = pd.read_csv('genome-scores.csv')

#genome_scores.groupby(['movieId', 'tagId']).apply(lambda x: x['relevance']).apply(pd.Series)


#genome_scores.loc[genome_scores['movieId']==1][:10]

genome_scores
```

```python
new_genome_scores = genome_scores.groupby('movieId').sample(10)
```

```python
df = pd.merge(movies, ratings, on = 'movieId', how = 'outer')
df = df.groupby(['movieId', 'title', 'genres']).mean().reset_index()


#df.loc[df['movieId'] == 61354]
```

```python
df.drop('userId', axis = 'columns', inplace = True)
```

```python
df2 = pd.merge(df, tags, on = 'movieId', how = 'outer')

df2.groupby(['movieId', 'title', 'genres', 'rating']).sample(10, replace = True)

df2
```

```python
df3 = df2.groupby(['movieId', 'title', 'genres', 'rating']).apply(lambda x: x['tag'].unique()).apply(pd.Series)

column_tag = []

for i in range(173):
    column_tag.append('tag' + str(i))
    
    

##we loose 20 movies due to sample
```

```python
df3.columns = column_tag
```

```python
df3.reset_index(inplace = True)
```

```python
df3.head()
```

```python
base = df3[['movieId', 'title', 'genres', 'rating']]

tag_flatten = df3[column_tag].fillna('No')

tag_flatten

```

```python
df4 = pd.merge(genome_tags, new_genome_scores, on = 'tagId', how = 'outer')

df4
```

```python
df5 = pd.merge(base, df4, on = 'movieId', how = 'outer')

col = ['movieId', 'title', 'genres', 'rating']

#df5.drop('tagId', axis = 'columns', inplace = True)

#df5.groupby(col).sample(10, replace = True)

df_pivot = pd.pivot_table(df5, values = ['relevance'], index = col, columns = ['genome'])

#df_pivot.reset_index().columns

#levels = df_pivot.columns.levels

#labels = df_pivot.columns.labels

#df_pivot.columns = levels[1][labels[1]]

#df_pivot

## fill nan with 0
```

```python
df_pivot.columns = df_pivot.columns.levels[1]
```

```python
df_pivot2 = df_pivot.reset_index()
```

```python
df_pivot3 = df_pivot2.fillna(0.000)
```

```python
df_pivot3[['genre_1', 'genre_2', 'genre_3', 'genre_4', 'genre_5', 'genre_6', 'genre_7', 'genre_8', 'genre_9', 'genre_10']] = df_pivot3['genres'].str.split('|', expand=True)

df_pivot3.drop('genres', axis='columns', inplace=True)

df_pivot3
```

```python
final_df = df_pivot3.join(tag_flatten)
```
